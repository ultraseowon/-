# 2주차 문제
## 정해진 숫자 맞추기 게임

```python
ans = 42
total = 1
# while 조건문:  조건문을 만족할 때까지 반복함.
while True:     # while True : 무한반복을 할 수 있게 함.
    a = int(input('숫자를 입력하세요 >>> '))    # 숫자를 입력받음.
    # if ~ elif ~ else 문을 사용함.
    # 조건이 여러 개일 때:연속 if문 사용. if 조건1:, elif 조건2:, ... , elif 조건n:, else:
    if a > ans:     # 답을 ans에 지정해놓은 다음 입력받음 a와 비교함.
        print('너무 큽니다.')
    elif a < ans:
        print('너무 작습니다.')
    else:
        print(f'정답입니다! 시도 횟수: {total}회')
        break       # break : 실행 중인 반복문을 중지
    total += 1      # a += b : 새로운 변수를 지정하지 않더라도 기존의 a라는 변수에 b를 더한 값을 a에 다시 지정해주는 할당 연산자.
    # while문이 한번 실핼될 때마다 숫자가 늘어 반복된 횟수를 알 수 있게 해줌.
```

# 3주차 문제

```python
# 문자열에서 특정 문자열만 출력하기
text = "level madam robot noon apple"
# .split() : 괄호 안의 문자열을 기준으로 기준으로 문자열을 나눠 리스트를 생성.
li = text.split(" ")
res = ""    # 빈 문자열 생성
for i in li: # i에 li의 단어를 하나씩 대입.
    if i == i[::-1]: # 슬라이싱 : [시작인덱스:종료인덱스+1:단계(이 간격으로 추출)], [::-1]의 경우 처음부터 끝까지 거꾸로 추출. 본 문자와 거꾸로 배열된 문자열을 비교.
        res += i + ' ' # 위의 조건 만족한 단어는 문자열에 추가. 이때 ' '는 단어를 구분하기 위한 띄어쓰기임.

print(res)
```

```python
# 별모양 출력하기
n = int(input('숫자 입력: '))
for i in range(1,n+1):  # for i in range(a, b, c): a부터 시작해서 b-1까지 c의 간격으로 i에 대입하여 그동안 반복. a, b, c는 모두 정수.
    for j in range(i):
        print('★', end='')     # 중첩 for문 : i가 1일 때, j에 1부터 1까지 대입. 1만큼 print 함수 반복. i가 3일 때, j에 1부터 3까지 대입. 3만큼 print 함수 반복. end=''를 통해 별 뒤에 공백 제거.
    print()     # 위의 실행 끝나면 줄바꿈.
    if i == n:  #  == 는 비교연산자, 결과 값은 논리형.
        for i in range(n-1,0,-1):   # i에 1부터 n까지 대입하다가 i가 n이 되면 n-1부터 0이 될 때까지 반복. 범위에는 양의 정수만 들어갈 수 있음. 거꾸로 반복.
            for j in range(i):
                print('★', end='')
            print()
```

```python
# 다이아 모양 출력하기
n = int(input('숫자 입력: '))
for i in range(1,n+1):
    # for문을 두개를 연달아 사용하면서 ' '와 별모양의 글자수의 합이 항상 n이 되게 설정.
    for j in range(n-i):
        print(' ', end='')  # i이 1일때 j에 1부터 n-2까지 대입하는 만큼 실행. 즉, n-1번 실행(?). n-1의 공백 출력
    for j in range(i):
        print('★', end='')
#    for j in range(n-i):
#        print(' ', end='')
    print()
for i in range(1,n+1):  # 1부터 n까지의 숫자만큼 반복.
    for j in range(i):
        print(' ', end='')
    for j in range(n-i):
        print('★', end='')
    print()

# 위, 아래 for문이 총 5번 실행이 되는데 아래의 반복문의 경우 처음 시작이 별 4개를 출력하는 것임.
# 만약 range(n-i)를 바꾼다면 절반씩은 피라미드식 삼각형이 출력이 될지는 몰라도 지금처럼 모양이 딱 맞는 다이아몬드의 형태는 출력되지 않음.
# 또는 별이 5개인 부분이 반복됨.
```

# 복습용 문제
- 숫자 2개와 연산자 하나를 입력받아 결과 값을 출력

```python
# 계산기
def calculator1(num1:int, num2:int, op:str):    # def 함수명(변수, 변수, 변수):  새로운 함수 생성. 변수 뒤 ':데이터형'을 입력하면 타입 힌트라고 하는 정보를 제공.
    if op == '+':                   # 이때 변수(?)뒤 :int를 적어주게 되면 해당 변수는 정수형만 입력받을 수 있다는 설명 추가 가능
        res = num1 + num2
        return res
                      # 컴퓨터에 res값 반환. 프린트 함수에서 출력하는 것과 반환은 다름. 반환은 사용자에게 보이지는 않지만 저장됨.
    elif op == '-':
        res = num1 - num2
        return res

    elif op == '*':
        res = num1 * num2
        return res

    else:
        res = num1 / num2
        return res
    print(res)

             # return은 출력을 해주지 않으므로 함수에 출력까지 하는 기능을 넣고 싶다면 작성.
```

```python
print(calculator1(20, 5, '+'))
print(calculator1(20, 5, '-'))
print(calculator1(20, 5, '*'))
print(calculator1(20, 5, '/'))

# 프린트 함수를 안 쓰면 마지막에 작성된 함수의 결과값만 출력이 되므로 모두 프린트를 작성
```
