## 문제 2
문자열 하나를 입력받아 다음을 출력하시오.
1. 전체 길이
2. 첫  글자
3. 마지막 글자
4. 앞 6글자
5. 뒤 4글자

- 입력 예시 : 991010-1234567
- 출력 예시 :
-       길이 : 14
        첫 글자 : 9
        마지막 글자 : 7
        앞 6글자 : 990101

핵심 개념

- len(), 인덱싱, 슬라이싱

````python
# 주민번호 느낌의 문자열 분석기
num = input('입력 >>> ')
print(
    f'길이 :, {len(num)}', f'첫 글자 : {num[0]}', f'마지막 글자 : {num[-1]}', f'앞 6글자 : {num[0:6]}', f'뒤 7글자 : {num[-4:]}', sep='\n'
    )

'''
    f-string 하나에 모든 정보를 다 넣고 줄 바꿈을 하고 싶었으나 실패하여 다 따로 입력.
    len()으로 길이
    num[0]으로 인덱싱. 왼쪽부터 0, 1, 2, ... 으로 인덱스 부여함. 따라서 [0]은 첫 글자. 오른쪽부터는 -1, -2, -3, ...으로 인덱스 부여. 따라서 [-1]은 마지막 글자.
    슬라이싱은 인덱싱 연속으로 여러개 한 것. [시작 인덱스:종료 인덱스+1]으로 지정. 이는 음수 인덱싱도 마찬가지. 숫자는 종료 인덱스가 더 크도록 설정. 끝까지 슬라이싱 할
    거면 공백으로 놔둠
    sep='' : 프린트 함수 안의 모든 데이터들을 괄호 안의 것을 기준으로 삼아 나누는 것.
    하나의 프린트 함수 내의 데이터들은 같은 줄에 공백 한칸 씩 출력되지만 sep=''로 분리 가능.
    \n : 줄 바꿈
'''
````

## 문제 3
문자열을 하나 입력받고, 그 안에 들어 있는 숫자 문자만 골라 모두 더한 값을 출력하시오.

- 입력 예시 : a2b3x9
- 출력 예시 : 합계 : 14

조건

- 문자열을 한 글자씩 검사할 것.
- 조건문을 사용할 것.

핵심 개념

- 문자열 순회, 조건문, 누적 합

````python
# 숫자만 모아서 합 구하기
sen = input('입력 >>> ')
res = 0     # 결과값을 받을 변수 생성
for num in sen:     # num에 sen을 한 글자씩 대입.
    if num in '0123456789':     # 문자열 순회 : 기준이 되는 문자열과 입력받은 문자열을 하나씩 비교∙대조해가며 검사.
        num = int(num)          # 위에서 비교한 데이터가 '문자열형'이므로 숫자임을 확인받은 글자(숫자)를 정수형으로 변환.
        res += num              # 만약 숫자라면 for문이 끝날 때까지 검사된 숫자들을 res에 더함.

print(f'합계 : {res}')
````

# 문제 9

````python
# 계산기 함수 만들기
def calc(a, op, b):
    match op:               # match case문 : 비교하고자 하는 입력값을 match에, 비교값을 case 뒤에 씀. match, case 줄의 마지막에 : 필수.
        case '+':           # case는 match의 한칸 옆에 씀. tab
             res = a+b
        case '-':
            res = a-b
        case '*':
            res = a*b
        case '/':
            res = a/b
        case _:             # case _ : 위에 제시된 조건 외의 다른 모든 조건. 반드시 case 뒤에 공백을 한 칸 띄우고 _ 입력.
            res = '잘못된 연산자입니다.'

    return res              # 결과값을 컴퓨터에 반환. return 이후에 작성된 코드는 실행 X


print(calc(2, '/', 7))
print(calc(2, '+', 9))
print(calc(3, '//', 6))
````

# 문제 10
학생 점수들이 담긴 리스트를 입력받았다고 가정하고, 다음 정보를 반환하는 함수를 작성하시오.
함수 이름 : analyze(scores)
반환해야 할 값 :
1. 총합
2. 평균
3. 최고점
4. 최저점

조건

- 함수 안에서 반복문을 사용해 직접 계산할 것
- max(), min(), sum() 없이 풀 것

핵심 개념

- 함수, 리스트 순회, 누적값, 비교





````python
def analyze(scores):
    sum = 0             # 합
    max = scores[0]     # 최댓값 임의 설정.
    min = scores[0]     # 최솟값    ||

    for score in scores:
        sum += score    # 총합 구하기
        if score > max:
            max = score # score에 새로운 값이 올 때마다 비교해서 더 큰 값이 있으면 교체.
        if score < min:
            min = score # score에 새로운 값이 올 때마다 비교해서 더 작은 값이 있으면 교체.
    avg = sum/len(scores) # 평균 구하기. 반복해서 구할 필요 X, for문에서 총합이 다 구해지면 구하기.
    return sum, avg, max, min

scores = [70, 85, 90, 60, 100]
sum, avg, max, min = analyze(scores) # 각각의 변수에 analyze()에서 return으로 반환받은 값들을 순서대로 대입.
print(f'총합 : {sum}')
print(f'평균: {avg}')
print(f'최고점 : {max}')
print(f'최저점 : {min}')
````
